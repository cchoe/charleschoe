```python
# Clean up text by stripping whitespace
# https://stackoverflow.com/questions/40950310/strip-trim-all-strings-of-a-dataframe
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
```

```python
# If a retailer doesn't have enough data points,
#  lump them in with other smaller retailers by channel
retailer_counts = df['retailer_name'].value_counts()
big_retailers = [r for r,n in retailer_counts.items() if n>10000] + ['Dollar General']
df['retailer big'] = df['retailer_name'].where(df['retailer_name'].isin(big_retailers),
                          df['channel_name'].apply(lambda x: 'Other ({})'.format(x)))
```

```python
df['transaction_dt'] = pd.to_datetime(df['transaction_dt'])
df['day of the week'] = df['transaction_dt'].apply(lambda x: x.weekday())
df['is_weekend'] = 1*(df['day of the week'] > 4)
```

```python
# df['matching city'] = 1*(df['user_geo_city'] == df['store_geo_city'])
```

```python
# Consider predictors based on user profile
# Dropping vars with too many levels (city, zip code)
# Dropping information about trip
user_profile_vars = ['user_id','user_geo_rural_code', 'user_geo_census_region',
       'user_geo_census_division', 'user_demo_ethnicity', 'user_demo_age', 'user_demo_education',
       'user_demo_employment', 'user_demo_gender',
       'user_demo_income', 'user_demo_marital_status',]
trip_vars = ['transaction_dt','day of the week','is_weekend']
```

```python
retailer_groups = df['retailer big'].unique()
df['retailer big'].value_counts(), df['retailer big'].value_counts().sum()
```

```python
# Predicting retailer
retailer_OHE = pd.get_dummies(df['retailer big'])
df_ohe = pd.concat([df[user_profile_vars+trip_vars],retailer_OHE], axis=1)
# Group by user, day
# gf = df_ohe.groupby(['user_id','transaction_dt'])
gf = df_ohe.groupby(user_profile_vars+trip_vars, as_index=False).max()
X_vars = ['user_geo_rural_code', 'user_geo_census_region',
        'user_geo_census_division', 'user_demo_ethnicity', 'user_demo_age', 'user_demo_education',
        'user_demo_employment', 'user_demo_gender',
        'user_demo_income', 'user_demo_marital_status',
          'day of the week','is_weekend']
gf_ohe = pd.get_dummies(gf[X_vars], drop_first=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
X_train, X_test = train_test_split(gf_ohe,random_state=0)
clfs = []
fis = []
reports = []
for ylabel in df['retailer big'].unique():
    print('Building RF classifier for ', ylabel)
    y_train, y_test = train_test_split(gf[ylabel],random_state=0)
    clf = RandomForestClassifier(max_depth=2, random_state=0, class_weight='balanced').fit(X_train, y_train)
    clfs.append(clf)
    fi = pd.Series(clf.feature_importances_, index=gf_ohe.columns)
    fis.append(fi)
    print('              precision    recall  f1-score   support')
    print(classification_report(y_test,clf.predict(X_test),labels=[1]).splitlines()[2])
    report = classification_report(y_test,clf.predict(X_test), output_dict=True)['1']
    reports.append(report)

# y_train, y_test = train_test_split(gf[retailer_groups],random_state=0)
# clf = RandomForestClassifier(max_depth=2, random_state=0, class_weight='balanced').fit(X_train, y_train)
# fi = pd.Series(clf.feature_importances_, index=gf_ohe.columns)
# clf.predict(X_test)
```

```python
# Clustering, see where Safeway looks most similar
from sklearn.cluster import KMeans
X = pd.get_dummies(df[['channel_name','user_geo_rural_code', 'user_geo_state', 'user_geo_census_region',
       'user_geo_census_division', 'payment_method',
       'basket_total_dollars_binned', 
       'user_demo_ethnicity', 'user_demo_age', 'user_demo_education', 'user_demo_gender',
       'user_demo_income', 'user_demo_marital_status',
       'kids_present_on_trip', 'adult_genders_on_trip',
       'were_there_other_stores_on_this_shopping_trip','is_weekend']],drop_first=True)
kmeans = KMeans(n_clusters=15, random_state=0).fit(X)
```
